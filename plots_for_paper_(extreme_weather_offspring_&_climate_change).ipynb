```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
mycolors = ['tab:green', 'tab:blue', 'tab:red']  
```

```python
#adding path to file
PATH_sim1 = "paper_resources/paper_extreme_weather_offspring.csv"
df_sim1 = pd.read_csv(PATH_sim1, comment='#')
df_sim1['Year'] = df_sim1.index
df_sim1
```

```python
#preparing data for plots
x_sim1  = df_sim1['Year'].values.tolist()

y1_sim1 = df_sim1['Spruce (count)'].values.tolist()
y2_sim1 = df_sim1['Fir (count)'].values.tolist()
y3_sim1 = df_sim1['Beech (count)'].values.tolist()
y_1_sim1 = np.vstack([y1_sim1, y2_sim1, y3_sim1])

y4_sim1 = df_sim1['Spruce volume (m^3)'].values.tolist()
y5_sim1 = df_sim1['Fir volume (m^3)'].values.tolist()
y6_sim1 = df_sim1['Beech volume (m^3)'].values.tolist()
y_2_sim1 = np.vstack([y4_sim1, y5_sim1, y6_sim1])
```

```python
#adding path to file
PATH_sim2 = "paper_resources/paper_climate_change.csv"
df_sim2 = pd.read_csv(PATH_sim2, comment='#')
df_sim2['Year'] = df_sim2.index
df_sim2
```

```python
#preparing data for plots
x_sim2  = df_sim2['Year'].values.tolist()

y1_sim2 = df_sim2['Spruce (count)'].values.tolist()
y2_sim2 = df_sim2['Fir (count)'].values.tolist()
y3_sim2 = df_sim2['Beech (count)'].values.tolist()
y_1_sim2 = np.vstack([y1_sim2, y2_sim2, y3_sim2])

y4_sim2 = df_sim2['Spruce volume (m^3)'].values.tolist()
y5_sim2 = df_sim2['Fir volume (m^3)'].values.tolist()
y6_sim2 = df_sim2['Beech volume (m^3)'].values.tolist()
y_2_sim2 = np.vstack([y4_sim2, y5_sim2, y6_sim2])
```

```python
#plots for first simulation
fig, ax = plt.subplots(figsize=(16, 9))
columns = df_sim1.columns[1:]
labs = columns.values.tolist()
ax = plt.gca()
ax.stackplot(x_sim1, y_1_sim1, labels=['Spruce','Fir','Beech'], colors=mycolors, alpha=0.8)
ax.set_title('Total number of trees (extreme weather and high offspring rate)', fontsize=18)
ax.legend(fontsize=10, ncol=3)
plt.xticks(x_sim1[::int((max(x_sim1)+1)/10)], fontsize=10, horizontalalignment='center')
plt.xticks(fontsize=10)
plt.xlim(x_sim1[0], x_sim1[-1])
ax.set_xlabel('Time (year)')
ax.set_ylabel('Number of trees (trunk)')
plt.savefig('paper_resources/paper_extreme_weather_offspring_count.png') 

fig, ax = plt.subplots(figsize=(16, 9))
columns = df_sim1.columns[1:]
labs = columns.values.tolist()
ax = plt.gca()
ax.stackplot(x_sim1, y_2_sim1, labels=['Spruce','Fir','Beech'], colors=mycolors, alpha=0.8)
ax.set_title('Total volume of trees (extreme weather and high offspring rate)', fontsize=18)
ax.legend(fontsize=10, ncol=3)
plt.xticks(x_sim1[::int((max(x_sim2)+1)/10)], fontsize=10, horizontalalignment='center')
plt.xticks(fontsize=10)
plt.xlim(x_sim1[0], x_sim1[-1])
ax.set_xlabel('Time (year)')
ax.set_ylabel('Volume of trees (m^3)')
plt.savefig('paper_resources/paper_extreme_weather_offspring_volume.png') 

#plots for second simulation
fig, ax = plt.subplots(figsize=(16, 9))
columns = df_sim2.columns[1:]
labs = columns.values.tolist()
ax = plt.gca()
ax.stackplot(x_sim2, y_1_sim2, labels=['Spruce','Fir','Beech'], colors=mycolors, alpha=0.8)
ax.set_title('Total number of trees (climate change)', fontsize=18)
ax.legend(fontsize=10, ncol=3)
plt.xticks(x_sim2[::int((max(x_sim2)+1)/10)], fontsize=10, horizontalalignment='center')
plt.xticks(fontsize=10)
plt.xlim(x_sim2[0], x_sim2[-1])
ax.set_xlabel('Time (year)')
ax.set_ylabel('Number of trees (trunk)')
plt.savefig('paper_resources/paper_climate_change_count.png') 

fig, ax = plt.subplots(figsize=(16, 9))
columns = df_sim2.columns[1:]
labs = columns.values.tolist()
ax = plt.gca()
ax.stackplot(x_sim2, y_2_sim2, labels=['Spruce','Fir','Beech'], colors=mycolors, alpha=0.8)
ax.set_title('Total volume of trees (climate change)', fontsize=18)
ax.legend(fontsize=10, ncol=3)
plt.xticks(x_sim2[::int((max(x_sim2)+1)/10)], fontsize=10, horizontalalignment='center')
plt.xticks(fontsize=10)
plt.xlim(x_sim2[0], x_sim2[-1])
ax.set_xlabel('Time (year)')
ax.set_ylabel('Volume of trees (m^3)')
plt.savefig('paper_resources/paper_climate_change_volume.png') 

plt.show()
```
